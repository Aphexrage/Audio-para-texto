```python
# Imports

from datasets import load_dataset
from pathlib import Path
import soundfile
import matplotlib.pyplot as plt
import IPython
import IPython.display
from datasets import Audio
```

```python
dataset = load_dataset('ashraq/esc50')
```

```python
dados = dataset['train']

primeirasLinhas = dados[0]
```

```python
primeirasLinhas = dados.select(range(10))
```

```python
for linha in primeirasLinhas:
    print(linha)
    print("---")
```

```python
idxDados = 1
linha = dados[idxDados]

plt.subplots(figsize=(30, 4))
plt.plot(linha['audio']['array'])
plt.suptitle(linha['category'])

plt.show()
```

```python
pastaSaida = Path('audios') / 'objetos'
pastaSaida.mkdir(exist_ok=True, parents=True)
```

```python


for i, linha in enumerate(primeirasLinhas):
    objeto = linha['category']
    dadosSom = linha['audio']['array']
    taxaAmostragem = linha['audio']['sampling_rate']
    caminhoSaida = pastaSaida / f'{i}_{objeto}.wav'
    #soundfile.write(file=caminhoSaida, data=dadosSom, samplerate=taxaAmostragem)
    display(IPython.display.Audio(data=dadosSom, rate= taxaAmostragem))
```
